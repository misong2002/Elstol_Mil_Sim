```python
Angelina = Intelligent(race=2,profession=2,level=2)
```

```python
sum(Angelina.attributes.values())
```

```python
weapon_db = WeaponDatabase("../data/weapon.csv")

# 获取特定武器
longsword = weapon_db.get_weapon(1)
if longsword:
    print(f"找到武器: {longsword}")
    
    # 示例属性值（9种属性）
    example_attributes = list(Angelina.attributes.values())[0:-1]
    
    # 计算穿刺伤害加成（索引0）
    pierce_bonus = longsword.calculate_total_damage_bonus(example_attributes, 0)
    print(f"穿刺伤害加成: {pierce_bonus}")
    
    # 计算挥砍伤害加成（索引1）
    slash_bonus = longsword.calculate_total_damage_bonus(example_attributes, 1)
    print(f"挥砍伤害加成: {slash_bonus}")

# 获取所有武器
all_weapons = weapon_db.get_all_weapons()
print(f"数据库中共有 {len(all_weapons)} 把武器")
```

```python
# 创建详细状态数据库
detailed_state_db = StateDatabase("../data/states.csv")

# 获取特定状态
flame_enchant_state = detailed_state_db.get_state(5)  # 火焰附魔
if flame_enchant_state:
    print(f"找到状态: {flame_enchant_state}")
    
    # 获取火焰魔法攻击修正
    fire_attack_mod = flame_enchant_state.get_magical_attack_modifier("fire")
    print(f"火焰魔法攻击修正: {fire_attack_mod}")

# 获取所有状态
all_states = detailed_state_db.get_all_states()
print(f"数据库中共有 {len(all_states)} 个状态")
```

```python
# 使用示例
if __name__ == "__main__":
    # 创建魔物数据库
    monster_db = MonsterDatabase("../data/monsters.csv")
    
    # 获取特定魔物
    aoshio = monster_db.get_monster(8)  # 火焰元素
    if aoshio:
        print(f"找到魔物: {aoshio}")
        print(f"描述: {aoshio.description}")
        # 获取火焰魔法攻击力
        fire_attack = aoshio.get_magical_attack("fire")
        print(f"奥术魔法攻击力: {fire_attack}")
        
        # 获取火焰防御力
        fire_defense = aoshio.get_magical_defense("fire")
        print(f"奥术防御力: {fire_defense}")
        
        # 获取恐惧力场信息
        print(f"恐惧力场强度: {aoshio.fear_strength}, 范围: {aoshio.fear_range}")
    
    # 获取所有大型魔物
    large_monsters = monster_db.get_monsters_by_size("large")
    print(f"数据库中共有 {len(large_monsters)} 个大型魔物")
    
    # 获取所有魔物
    all_monsters = monster_db.get_all_monsters()
    print(f"数据库中共有 {len(all_monsters)} 个魔物")
```
